### Running Proportion

```python
# imports
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# set constants
nFlips = 500
pHeads = 0.5
```

```python
# generate sequence
nHeads = np.random.choice([0,1], nFlips, replace=True)

# Compute the running proportion of heads:
r = np.cumsum(nHeads)
n = np.linspace(1, nFlips, nFlips)  # n is a vector.
run_prop = r/n  # component by component division.
```

```python
nHeads[:10]
```

```python
run_prop[:10]
```

```python
# plot the results
plt.scatter(x=range(nFlips), y=run_prop)
plt.plot(range(nFlips), run_prop)
plt.xlim(0, nFlips)
plt.ylim(0, 1)
plt.title("Runing Proportion of heads")
plt.ylabel("Proportion heads")
plt.xlabel("Flip number")
plt.ylabel("Proportion heads")

# Display the beginning of the flip sequence.
flipletters = ''.join(["T","H"][flip] for flip in nHeads[:10])

plt.annotate('Flip Sequence = ' + flipletters + '...' , xy=(150, 0.8) )

plt.annotate('final proportion\n' + str(run_prop[-1]), xy=(nFlips-150, run_prop[-1]-0.2))
plt.show()
```

### Area under the Curve

```python
mu = 5.0
sd = 2.0
delta = 0.2


def normval(x):
    return (1/(sd * np.sqrt(2*math.pi))) * np.exp(-1/(2*sd**2)*(x-mu)**2)


x = np.arange(mu-3*sd, mu+3*sd, delta)
y = [normval(i) for i in x]
total = round(delta*sum(y), 6)

plt.plot(x, y)
plt.stem(x, y, markerfmt=' ')

plt.title("normal density plot")
plt.xlabel("x")
plt.ylabel("p(x)")
plt.annotate('$\mu$ = {m}\n$\sigma$ = {s}'.format(m=mu, s=sd), xy=(mu-4, np.max(y)*0.8))
plt.annotate('$\Sigma\Delta$p(x) = {t}'.format(t=total), xy=(mu+2, np.max(y)*0.8))
plt.show()
```
